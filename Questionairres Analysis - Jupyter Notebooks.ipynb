```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("survey.csv")
df.head()
```

```python
group_Corr = pd.crosstab(df["What is your gender?"], df["Do you typically go out in a group or alone?"])
group_Corr
```

```python
group_Corr.plot.bar()
```

```python
df2 = df[(df["What is your gender?"]=='Male')]
ManGender = df2.groupby("What is your gender?")["Do you typically go out in a group or alone?"].value_counts().plot.bar()
ManGender
```

```python
df3 = df[(df["What is your gender?"]=='Female')]
FemGender = df3.groupby("What is your gender?")["Do you typically go out in a group or alone?"].value_counts().plot.bar()
FemGender
```

```python
genderscross = pd.crosstab(df["What is your gender?"], df["Do you typically go out in a group or alone?"],
            normalize=True, margins=True)
genderscross
```

```python
group_Corr.plot.bar(stacked=True)
```

```python
df4 = df[(df["What is your gender?"]=='Male')]
MalePie = df4.groupby("What is your gender?")["Do you typically go out in a group or alone?"].value_counts().plot.pie()
```

```python
df5 = df[(df["What is your gender?"]=='Female')]
FemalePie = df5.groupby("What is your gender?")["Do you typically go out in a group or alone?"].value_counts().plot.pie()
```
